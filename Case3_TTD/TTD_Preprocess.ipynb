```python
import pandas as pd
import numpy as np
```

```python
def split_line(string):
    string = string.replace('\n','')
    L_list = string.split("\t")
    while '' in L_list:
        L_list.remove('')
    return L_list
```

# load data

# targets

```python
f = open("/data3/cyx/TTD/P1-01-TTD_target_download.txt")
line_list = f.readlines()
```

```python
line_list[:40]
```

```python
line_list = line_list[40:]
```

```python
line_list_processed = [split_line(x) for x in line_list]
```

```python
len(line_list_processed)
```

```python
df_target = pd.DataFrame(line_list_processed)
df_target = df_target.loc[- pd.isnull(df_target.loc[:,0]),:]
df_target.columns = ["target_id","property","value1","value2","value3"]
```

```python
target_list = pd.unique(df_target["target_id"])
print(str(len(target_list)) + " targets")
```

```python
gene_list = pd.unique(df_target.loc[df_target["property"]=="GENENAME","value1"])
print(str(len(gene_list)) + " genes")
```

```python
df_target
```

```python
def simple_target(x):
    df_x = df_target.loc[df_target.target_id==x,]
    list_x = dict()
    list_x["TARGETID"] = x
    for col in  ["TARGNAME","GENENAME",'FORMERID','UNIPROID',"TARGTYPE","SYNONYMS",'FUNCTION','PDBSTRUC','BIOCLASS']:
        element = df_x.loc[df_x["property"]==col,"value1"].tolist()
        if len(element)>0:
            list_x[col] = element[0]
        else:
            list_x[col] = None
    return list_x
```

```python
df_target_trans = pd.DataFrame([simple_target(x) for x in target_list])
df_target_trans.columns = ["TARGETID","TARGNAME","GENENAME",'FORMERID','UNIPROID',"TARGTYPE","SYNONYMS",'FUNCTION','PDBSTRUC','BIOCLASS']
```

```python
df_target_trans
```

```python
df_target.to_csv("raw.targets.csv")
```

```python
df_target_trans.to_csv("processed.targets.csv")
```

## drug-target pair

```python
df_target_trans = pd.read_csv("TTD/processed.targets.csv",index_col=0)
df_drug_trans = pd.read_csv("TTD/processed.drug.csv",index_col=0)
```

```python
df_pair = pd.read_excel("/data3/cyx/TTD/P1-07-Drug-TargetMapping.xlsx")
```

```python
df_pair
```

```python
target_intersection = list(set(df_pair.TargetID) & set(df_target_trans.TARGETID))
drug_intersection = list(set(df_pair.DrugID) & set(df_drug_trans.DRUG__ID))
```

```python
len(target_intersection)
```

```python
len(drug_intersection)
```

```python
df_pair = df_pair.loc[[(x in target_intersection) for x in df_pair.TargetID],:]
df_pair = df_pair.loc[[(x in drug_intersection) for x in df_pair.DrugID],:]
```

```python
df_pair
```

```python
df_pair.loc[:,"TargetName"] = [df_target_trans.loc[df_target_trans.TARGETID == x,"TARGNAME"].values[0] for x in df_pair.TargetID]
```

```python
df_pair.loc[:,"TargetGene"] = [df_target_trans.loc[df_target_trans.TARGETID == x,"GENENAME"].values[0] for x in df_pair.TargetID]
```

```python
df_pair.loc[:,"DrugName"] = [df_drug_trans.loc[df_drug_trans.DRUG__ID == x,"DRUGNAME"].values[0] for x in df_pair.DrugID]
```

```python
df_pair
```

```python
df_pair.to_csv("processed.mapping.csv")
```
