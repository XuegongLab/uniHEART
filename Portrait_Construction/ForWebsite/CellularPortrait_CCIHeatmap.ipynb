```r
library(rjson)
library(reshape)
```

```r
cell_types <- c('Endothelial cell','Cardiomyocyte cell','Myeloid cell','Fibroblast','Lymphoid cell','Smooth muscle cell','Neuron','Pericyte','Adipocyte')
```

# Adult

```r
subregions <- c("LV", "IVS", "RV", "LA", "IAS", "RA")
```

```r
for(sr in subregions){
    dir.create(paste0("/stor/public/chenyx/HHCAd/ForWebsite220604//HeatMap_CCI/Adult/",sr),recursive = TRUE)
}
```

```r
for(sr in subregions){
    for(ct in cell_types){
        # load data
        path <- paste0("~/HHCA_d/Code4Paper/cell-cell interaction/Adult.Result/Heatmap//",sr,"/",ct,".csv")
        df <- read.csv(path)
        df$X <- c("Adipocyte", "Cardiomyocyte cell", "Endothelial cell", "Fibroblast", "Lymphoid cell", "Myeloid cell", "Neuron", "Pericyte", "Smooth muscle cell")
        mat <- melt(df)


        # write x ais
        json_str <- '{"xAxis": ["Adipocyte", "Cardiomyocyte cell", "Endothelial cell", "Fibroblast", "Lymphoid cell", "Myeloid cell", "Neuron", "Pericyte", "Smooth muscle cell"], '
        # write y axis
        json_str <- paste0(json_str, '"yAxis": [')
        for(gene in colnames(df)[2:length(colnames(df))]){
            json_str <- paste0(json_str, '"', gene ,'", ')
        }
        json_str <- substr(json_str,1,nchar(json_str)-2)
        json_str <- paste0(json_str, '], ')
        # write data
        json_str <- paste0(json_str, '"data": [')
        for (i in 1:dim(mat)[1]){
            json_str <- paste0(json_str,'[', '"',mat[i,1],'", "', mat[i,2],'", "',mat[i,3], '"],')
        }
        json_str <- substr(json_str,1,nchar(json_str)-1)
        json_str <- paste0(json_str, ']}')
        # save json file
        writeLines(json_str,paste0("/stor/public/chenyx/HHCAd/ForWebsite220604//HeatMap_CCI/Adult/",sr,"/",ct,".json"))
    }
}
```

# Fetal

```r
subregions <- c("LV", "IVS", "RV", "LA", "RA", "AO", "PA", "MV", "TV")
```

```r
for(sr in subregions){
    dir.create(paste0("/stor/public/chenyx/HHCAd/ForWebsite220604//HeatMap_CCI/Fetal/",sr),recursive = TRUE)
}
```

```r
for(sr in subregions){
    sr.path <- paste0("~/HHCA_d/Code4Paper/cell-cell interaction/Fetal.Result//Heatmap//",sr)
    cell_types <- as.character(lapply(dir(sr.path),function(x){strsplit(x,"[.]")[[1]][1]}))
    for(ct in cell_types){
        # load data
        path <- paste0("~/HHCA_d/Code4Paper/cell-cell interaction/Fetal.Result//Heatmap//",sr,"/",ct,".csv")
        df <- read.csv(path)
        df$X[ df$X=="Lymphocyte"] <- "Lymphoid cell"
        df$X[ df$X=="Myelocyte"] <- "Myeloid cell"
        
        mat <- melt(df)


        # write x ais
        json_str <- '{"xAxis": ["Adipocyte", "Cardiomyocyte cell", "Endothelial cell", "Fibroblast", "Lymphoid cell", "Myeloid cell", "Neuron", "Pericyte", "Smooth muscle cell"], '
        # write y axis
        json_str <- paste0(json_str, '"yAxis": [')
        for(gene in colnames(df)[2:length(colnames(df))]){
            json_str <- paste0(json_str, '"', gene ,'", ')
        }
        json_str <- substr(json_str,1,nchar(json_str)-2)
        json_str <- paste0(json_str, '], ')
        # write data
        json_str <- paste0(json_str, '"data": [')
        for (i in 1:dim(mat)[1]){
            json_str <- paste0(json_str,'[', '"',mat[i,1],'", "', mat[i,2],'", "',mat[i,3], '"],')
        }
        json_str <- substr(json_str,1,nchar(json_str)-1)
        json_str <- paste0(json_str, ']}')
        # save json file
        writeLines(json_str,paste0("/stor/public/chenyx/HHCAd/ForWebsite220604//HeatMap_CCI/Fetal/",sr,"/",ct,".json"))
    }
}
```
