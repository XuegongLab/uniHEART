```r
library(rjson)
library(reshape)
```

```r
cell_types <- c('Endothelial cell','Cardiomyocyte cell','Myeloid cell','Fibroblast','Lymphoid cell','Smooth muscle cell','Neuron','Pericyte','Adipocyte')
```

# Adult

```r
subregions <- c("LV", "IVS", "RV", "LA", "IAS", "RA")
```

```r
for(sr in subregions){
    dir.create(paste0("/stor/public/chenyx/HHCAd/ForWebsite220604//ChordPlot_CCI/Adult/",sr),recursive = TRUE)
}
```

```r
for(sr in subregions){
    
    path_sr <- paste0("~/HHCA_d/Code4Paper/cell-cell interaction/Adult.Result/Chordplot//",sr,"/")
    pathways <- lapply(strsplit(dir(path_sr),'[.]'),function(x){x[1]})
    for(pw in pathways){
        path <- paste0(path_sr,pw,".csv")
        # load data
        df <- read.csv(path)
        df$X[df$X=="Lymphocyte"] <- "Lymphoid cell"
        df$X[df$X=="Myelocyte"] <- "Myeloid cell"
        
        colnames(df)[colnames(df)=="Lymphocyte"] <- "Lymphoid cell"
        colnames(df)[colnames(df)=="Myelocyte"] <- "Myeloid cell"
        colnames(df)[colnames(df)=="Cardiomyocyte.cell"] <- "Cardiomyocyte cell"
        colnames(df)[colnames(df)=="Endothelial.cell"] <- "Endothelial cell"
        colnames(df)[colnames(df)=="Smooth.muscle.cell"] <- "Smooth muscle cell"
        mat <- melt(df)


        json_str <- '['
        for (i in 1:dim(mat)[1]){
            if(mat[i,3]>0.1){
                json_str <- paste0(json_str,'{"source": "',mat[i,1],'", "target": "', mat[i,2],'","value": ',mat[i,3]*10, '},')
            }
        }
        json_str <- substr(json_str,1,nchar(json_str)-1)
        json_str <- paste0(json_str, '] ')
        
        # save json file
        writeLines(json_str,paste0("/stor/public/chenyx/HHCAd/ForWebsite220604/ChordPlot_CCI/Adult/",sr,"/",pw,".json"))
    }
}
```
