```r
library(stringr)
library(ggplot2)
library(reshape2)
```

# Adult

```r
subregion.list <- c("LV", "IVS", "RV", "LA", "IAS", "RA")
```

```r
for(sr in subregion.list){
    path.sr <- paste0("/data1/chenyx/HHCAd/coexpression/Adult/", sr,"/")
    file.list <- dir(path.sr)
    for(file in file.list){
        df <- read.csv(paste0(path.sr, file),row.names = 1)
        ct <- strsplit(file,"[.]")[[1]][1]
        genes <- row.names(df)
        for(g in genes){
            gene_select <- c(g,genes[order(df[rownames(df)==g],decreasing = T)<=10])
            df_g <- df[rownames(df)%in% gene_select,rownames(df)%in% gene_select]
            df_g <- round(df_g*100,2)

            # write nodes
            json_str <- '{"nodes":['
            for(i in 1:11){
                json_str <- paste0(json_str,"{",
                                  '"id": ', '"',i-1,'",',
                                  '"name": ','"',gene_select[i],'",',
                                  '"category": ')
                if(i==1){
                    json_str <- paste0(json_str,'0},')
                }else{
                    json_str <- paste0(json_str,'1},')
                }
            }
            json_str <- substr(json_str,1,nchar(json_str)-1)
            json_str <- paste0(json_str,'],')
            # write edges
            json_str <- paste0(json_str,'"links": [ ')
            for(i in 2:11){
                for(j in 1:(i-1)){
                    json_str <- paste0(json_str,"{",
                                  '"source": ', '"',i-1,'",',
                                  '"target": ','"',j-1,'",',
                                  '"value": ','"',df_g[i,j],'"},')
                }
            }
            json_str <- substr(json_str,1,nchar(json_str)-1)
            json_str <- paste0(json_str,'],')
            # write categories
            json_str <- paste0(json_str,'"categories": [ {"name": "selected gene"}, {"name": "related genes"}]}')

            # save json
            dir.create(paste0("/data3/cyx/ForDatabase0419/Network/Adult/",sr,'/',ct),recursive = T)
            writeLines(json_str,paste0("/data3/cyx/ForDatabase0419/Network/Adult/",sr,'/',ct,'/',g,".json"))
        }
    }
}
```
