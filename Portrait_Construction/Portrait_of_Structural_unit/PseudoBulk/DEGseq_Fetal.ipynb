## Build data

```r
df_IVS <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk_fetal/Interventricular septum.csv",row.names = 1)

df_LV <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk_fetal/Left ventricle.csv",row.names = 1)
df_LA <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk_fetal/Left atria.csv",row.names = 1)

df_RV <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk_fetal/Right ventricle.csv",row.names = 1)
df_RA <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk_fetal/Right atria.csv",row.names = 1)

df_AO <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk_fetal/Aorta.csv",row.names = 1)
df_PA <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk_fetal/Pulmonary artery.csv",row.names = 1)

df_MV <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk_fetal/Mitral valve.csv",row.names = 1)
df_TV <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk_fetal/Tricuspid valve.csv",row.names = 1)
```

```r
colnames(df_IVS) <- paste0("IVS_",1:50)
colnames(df_LV) <- paste0("LV_",1:50)
colnames(df_LA) <- paste0("LA_",1:50)
colnames(df_RV) <- paste0("RV_",1:50)
colnames(df_RA) <- paste0("RA_",1:50)
colnames(df_AO) <- paste0("AO_",1:50)
colnames(df_PA) <- paste0("PA_",1:50)
colnames(df_MV) <- paste0("MV_",1:50)
colnames(df_TV) <- paste0("TV_",1:50)
```

```r
df <- cbind(df_LV,df_IVS,df_RV,df_LA,df_RA,df_AO,df_PA,df_MV,df_TV)
```

```r
saveRDS(df,"/stor/public/chenyx/HHCAd/PseudoBulk_fetal/merged.rds")
```

## process data

```r
df <- readRDS("/stor/public/chenyx/HHCAd/PseudoBulk_fetal/merged.rds")
```

```r
df <- df[rowMeans(df)>1,]
```

```r
df <- df/colSums(df)*10000
df <- log1p(df)
```

```r
label <- c(rep("LV", times=50),rep("IVS", times=50),rep("RV", times=50),rep("LA", times=50),rep("RA", times=50),rep("AO", times=50),rep("PA", times=50),rep("MV", times=50),rep("TV", times=50))
```

```r
df
```

# DEGseq 

```r
unit_list <-c("LV","IVS","RV","LA","RA","AO","PA","MV","TV")
```

```r
for (unit in unit_list){
    genes <- rownames(df)
    mat1 <- cbind(genes,df[,label==unit])
    mat1 <- as.matrix(mat1)

    mat2 <- cbind(genes,df[,!label==unit])
    mat2 <- as.matrix(mat2)
    
    outputDIr <- paste0("/stor/public/chenyx/HHCAd/DEGene_anatomical_fetal/",unit)
    
    
    DEGexp(mat1,geneCol1 = 1,expCol1 = 2:dim(mat1)[2],groupLabel1 = unit,
       geneExpMatrix2 = mat2,geneCol2 = 1,expCol2 = 2:dim(mat2)[2],groupLabel2 = "others",
       method = "LRT",pValue=0.05,outputDir = outputDIr)
}
```
