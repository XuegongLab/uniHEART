```r
library(org.Hs.eg.db)
library(stringr)
library(edgeR)
library(DEGseq)
```

# Subregion

## Build data

```r
df_IVS <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk/Interventricular septum.csv",row.names = 1)
df_IAS <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk/Interatrial septum.csv",row.names = 1)

df_LV <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk/Left ventricle.csv",row.names = 1)
df_LA <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk/Left atria.csv",row.names = 1)


df_RV <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk/Right ventricle.csv",row.names = 1)
df_RA <- read.csv("/stor/public/chenyx/HHCAd/PseudoBulk/Right atria.csv",row.names = 1)
```

```r
colnames(df_IVS) <- paste0("IVS_",1:50)
colnames(df_IAS) <- paste0("IAS_",1:50)
colnames(df_LV) <- paste0("LV_",1:50)
colnames(df_LA) <- paste0("LA_",1:50)
colnames(df_RV) <- paste0("RV_",1:50)
colnames(df_RA) <- paste0("RA_",1:50)
```

```r
df <- cbind(df_IAS,df_IVS,df_LV,df_LA,df_RV,df_RA)
```

```r
saveRDS(df,"/stor/public/chenyx/HHCAd/PseudoBulk/merged.rds")
```

## process data

```r
df <- readRDS("/stor/public/chenyx/HHCAd/PseudoBulk/merged.rds")
```

```r
df <- df[rowMeans(df)>1,]
```

```r
df <- df/colSums(df)*10000
df <- log1p(df)
```

```r
df
```

```r
label <- c(rep("IAS", times=50),rep("IVS", times=50),rep("LV", times=50),rep("LA", times=50),rep("RV", times=50),rep("RA", times=50))
```

## DEGseq 

```r
unit_list <-c("IAS","IVS","LV","LA","RV","RA")
```

```r
for (unit in unit_list){
    genes <- rownames(df)
    mat1 <- cbind(genes,df[,label==unit])
    mat1 <- as.matrix(mat1)

    mat2 <- cbind(genes,df[,!label==unit])
    mat2 <- as.matrix(mat2)
    
    outputDIr <- paste0("/stor/public/chenyx/HHCAd/DEGene_anatomical/",unit)
    
    
    DEGexp(mat1,geneCol1 = 1,expCol1 = 2:dim(mat1)[2],groupLabel1 = unit,
       geneExpMatrix2 = mat2,geneCol2 = 1,expCol2 = 2:dim(mat2)[2],groupLabel2 = "others",
       method = "LRT",pValue=0.05,outputDir = outputDIr)
}
```
